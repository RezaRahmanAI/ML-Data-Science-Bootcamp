# Which one should you use? (pyplot or Matplotlib OO method)

* When plotting something quickly, okey to use the pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = hd[hd.age > 50]
over_50
```

```python
len(over_50)
```

```python
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target')
```

```python
# OO Method

fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax)

ax.set_xlim([45, 80])
```

```python
fix, ax = plt.subplots(figsize = (10, 6))

over_50.plot(kind = 'scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
```

```python
over_50.target.values
```

```python
fix, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target']
)

ax.set(
    title='Heart Disease and Cholestrol Lebels',
    xlabel='Ages',
    ylabel='Cholestrol'
)

# Adding legend
ax.legend( *scatter.legend_elements(), title='Target')

# Adding mean() horizontal line
ax.axhline(over_50['chol'].mean() , linestyle='--')
```

```python
# Now plot on gender
fix, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(
    x=over_50.age,
    y=over_50.chol,
    c=over_50.sex
)

ax.legend( *scatter.legend_elements(), title = 'Sex')

ax.set(
    title='Heart Disease and Cholestrol Labels',
    xlabel='Age',
    ylabel='Cholestrol'
)

ax.axhline(over_50.chol.mean())
```

```python
over_50
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(
    nrows = 2,
    ncols = 1,
    figsize = (10,10),
    sharex=True,
)


scatter = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target)

ax0.axhline(over_50.chol.mean(), linestyle='--')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.set(title='Hear Disease and Cholestrol Lebels',
        #xlabel='Age',
        ylabel='Cholestrol')



scatter = ax1.scatter(x=over_50.age,
                     y=over_50.thalach,
                     c=over_50.target)

ax1.axhline(over_50.thalach.mean(), linestyle='--')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.set(title='Heart Disease and Thalach Lavels',
        xlabel='Age',
        ylabel='Thalach'
)


fig.suptitle('Heart Disease Analysis', fontsize=15, fontweight='bold')
```
