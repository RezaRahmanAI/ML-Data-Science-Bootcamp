# Make sure all data are numerical

```python
car_data = pd.read_csv('Data/car-sales-extended.csv')
car_data.shape, len(car_data)
```

```python
car_data.head()
```

```python
x = car_data.drop('Price', axis=1)
y = car_data.Price
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

# Creating model

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
x.head()
```

```python
# Turn categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

cat_features = ['Make', 'Colour']
one_hot = OneHotEncoder()
# transformer = ColumnTransformer(
#     [(
#         'one_hot',
#         one_hot,
#         cat_feature,
#     )],
#     remainder = 'passthrough'
# )

transformer = ColumnTransformer(
    transformers=[
        ('one_hot', one_hot, cat_features)],
    remainder='passthrough'
)

# Fit and transform your input data 'x'
transformed_x = transformer.fit_transform(x)
```

```python
transformed_x
```

```python
transformed_x = pd.DataFrame(transformed_x)
transformed_x
```

```python
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)
model.fit(x_train, y_train,)
model.score(x_test, y_test)
```

# Missing Data

```python
cmd = pd.read_csv('Data/car-sales-extended-missing-data.csv')
cmd
```

```python
cmd.isna().sum()
```

```python
cmd['Odometer (KM)'].fillna(cmd['Odometer (KM)'].mean(), inplace= True)

cmd['Doors'].fillna(4, inplace= True)
cmd['Make'].fillna('Missing', inplace = True)
cmd['Colour'].fillna('Missing', inplace = True)
```

```python
cmd.isna().sum()
```

```python
cmd.dropna(inplace = True)
```

```python
cmd.isna().sum()
```

```python
x = cmd.drop('Price', axis = 1)
y = cmd['Price']


cat_features = ['Make', 'Colour']
one_hot = OneHotEncoder()
# transformer = ColumnTransformer(
#     [(
#         'one_hot',
#         one_hot,
#         cat_feature,
#     )],
#     remainder = 'passthrough'
# )

transformer = ColumnTransformer(
    transformers=[
        ('one_hot', one_hot, cat_features)],
    remainder='passthrough'
)

# Fit and transform your input data 'x'
transformed_x_train = transformer.fit_transform(x)
# x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(transformed_x_train, y, test_size=0.2)
model.fit(x_train, y_train,)
model.score(x_test, y_test)
```
