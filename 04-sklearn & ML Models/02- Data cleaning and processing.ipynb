# Choosing the right estimator/algorithm for our specific problem

Some things to note:
   * Sklearn refers to machine learning model, algorithms as estimators.
   * Classification problem - predicting a category (heart diseases or not)
   * Regression problem - predicting a number (selling price of a car)

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
home_df = pd.DataFrame(housing['data'], columns = housing['feature_names'])
home_df
```

```python
home_df['Target'] = housing['target']
home_df
```

```python
np.random.seed(42)
# Import Algorithm
from sklearn.linear_model import Ridge

# Create Data
x = home_df.drop('Target', axis=1)
y = home_df.Target

# splitting
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)

# Fitting
model = Ridge()
model.fit(x_train, y_train)

# Scorring
model.score(x_test, y_test)
```
